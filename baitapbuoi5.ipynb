```python
#W5A1

def max(a: int, b: int) -> int:
  if a > b:
    return a
  else:
    return b

a, b = map(int, input().split())
print(max(a, b))
```

```python
#W5A2

def swap(a, b):
  return b, a

a, b = map(int, input().split())
print(*swap(a, b))
```

```python
#W5A3

import math

def prime(n):
  if n == 2 or n == 3:
    return True
  elif n <= 1 or n % 2 == 0:
    return False
  else:
    for i in range(3, math.isqrt(n) + 1, 2):
      if n % i == 0:
        return False
    return True

n = int(input())
print(prime(n))
```

```python
#W5A4

def perfect(n: int):
  tong = 0
  for i in range(1, n):
    if n % i == 0:
      tong += i
  if tong == n:
    return True
  else:
    return False

n = int(input())
print(perfect(n))
```

```python
#W5A5

def whatever(s: list, k: int):
  for idx, val in enumerate(s, start = 1):
    if val == k:
      return idx
  return -1

s = list(map(int, input().split()))
k = int(input())
print(whatever(s, k))
```

```python
#W5A6

def giaithua(n: int) -> int:
  tich = 1
  for i in range(n, 1, -1):
    tich *= i
  return tich

n = int(input())
print(giaithua(n))
```

```python
#W5A8

def hamming(a: int, b: int) -> int:
  return bin(a ^ b).count("1")

#cach2

def hamming(a: int, b: int) -> int:
  x = bin(a)[:2]
  y = bin(b)[:2]
  lenmax = max(len(x), len(y))
  x = x.zfill(lenmax)
  y = y.zfill(lenmax)
  cnt = 0
  for v, t in zip(x, y):
    if v == t:
      cnt += 1
  return cnt

a, b = map(int, input().split())
print(hamming(a, b))
```

```python
#W5A9

def tong(n: str) -> int:
  tong = 0
  for i in n:
    tong += int(i)
  return tong

n = input()
print(tong(n))
```

```python
#W5A10

def iso(a, b):
  mapatob = {}
  bused = set()
  for chara, charb in zip(a, b):
    if chara in mapatob:
      if mapatob[chara] != charb:
        return False
    else:
      if charb in bused:
        return False
      mapatob[chara] = charb
      bused.add(charb)
  return True

a, b = input().split()
print(iso(a, b))
```
