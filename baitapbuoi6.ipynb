```python
#W6A4

a = input()

b = {}

pairs = a.split()

for pair in pairs:
    if ":" in pair:
        key, value = pair.split(":", 1)
        c = b.get(key, [])
        c.append(value)
        b[key] = c

print(b)
```

```python
#W6A5

a = list(map(int, input().split()))

b = {"positives": 0, "negatives": 0, "zeros": 0}

for each in a:
    if each > 0:
        b["positives"] += 1
    elif each == 0:
        b["zeros"] += 1
    else:
        b["negatives"] += 1

print(b)
```

```python
#W6A9

import ast
from collections import Counter

a = input()
b = input()

aaf = ast.literal_eval(a)
baf = ast.literal_eval(b)

c = Counter(aaf) + Counter(baf)

d = dict(sorted(c.items()))

print(d)
```

```python
#W6A10

a = list(map(int, input().split()))
k = int(input())

b = set()

for i in range(len(a)):
    for e in range(i + 1, len(a)):
        if a[e] + a[i] == k:
            b.add((a[i], a[e]))
            break

print(list(sorted(b)))
```

```python
#W6A12

from collections import Counter

a = list(map(int, input().split()))

b = Counter(a)

c = [key for key, value in b.items() if value == max(b.values())]

print(min(c))
```

```python
#W6A15

a = input().split()
k = int(input())

key = a[::2]
value = a[1::2]

b = {
    x: y for x, y in zip(key, value) if y > k
}

print(b)
```

```python
#W6A19

def maxx(n):
    return max(n)

a = list(map(int, input().split()))
print(maxx(a))
```

```python
#W6A20

def wev(n: list, k: int) -> int:
    for i, val in enumerate(n):
        if val == k:
            return i + 1
    return -1

a = list(map(int, input().split()))
k = int(input())
print(wev(a, k))
```

```python
#W6A21

a = input().split()

b = {}

for i, word in enumerate(a):
    c = b.setdefault(word, ())
    b[word] = c + (i,)

print(b)
```
