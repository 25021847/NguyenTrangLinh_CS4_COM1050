```python
#W10A1

try:
  a, b = map(int, input().split())
  print(f"{(a / b):.2f}")
except ZeroDivisionError:
  print("Loi chia khong")
```

```python
#W10A2

try:
  s = list(map(int, input().split()))
  n = len(s)
  print(s[n])
except IndexError:
  print("Loi truy cap chi so")
```

```python
#W10A3

def is_happy(s):
  seen = set()
  while s != 1 and s not in seen:
    seen.add(s)
    s = sum(int(digit) ** 2 for digit in str(s))
  return s == 1

try:
  s = input()
  a = int(s)
  if is_happy(a):
    print("YES")
  else:
    print("NO")
except ValueError:
  print("Loi du lieu dau vao")
```

```python
#W10A4

try:
  a, b = map(int, input().split())
except ValueError:
  print("Loi nhap lieu")
else:
  print(a + b)
finally:
  print("Ket thuc chuong trinh")
```

```python
#W10A5

mix_list = ["10", "20", "abc", "30", "4.5", "50"]
int_list = []

for item in mix_list:
    try:
        num = int(item)
        int_list.append(num)
    except ValueError:
        print(f"Cảnh báo: Phần tử '{item}' không thể chuyển sang int.")

print("List sau khi chuyển:", int_list)
```

```python
#W10A6

import math

try:
  a = float(input())
  if a < 0:
    raise ValueError
  print(f"{math.sqrt(a):.2f}")
except ValueError:
  print("So am")
```
