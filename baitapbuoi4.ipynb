```python
#W4A2

import math

while True:
  n = int(input())
  prime = True
  if n <= 1:
    if n < 0:
      print("Nhập số liệu phù hợp.")
      continue
    else:
      prime = False
  elif n == 2:
    prime = True
  elif n % 2 == 0:
    prime = False
  else:
    i = 3
    while i <= math.sqrt(n):
      if n % i == 0:
        prime = False
        break
      i += 2

  if prime:
    print("True")
  else:
    print("False")

  break
```

```python
#W4A6

import math

def prime(n):
  if n <= 1:
    return False
  elif n == 2:
    return True
  elif n % 2 == 0:
    return False
  else:
    i = 3
    while i <= math.sqrt(n):
      if n % i == 0:
        return False
      i += 2
    return True

while True:

  a, b = map(int, input().split())
  tong = 0
  for i in range(a, b+1):
    if prime(i):
      tong += i

  print(tong)
  break
```

```python
#W4A7

import math

def prime(n):
  if n < 2:
    return False
  else:
    for i in range(2, int(math.sqrt(n))+1):
      if n % i == 0:
        return False
    return True

while True:

  n = int(input())

  for i in range(n, 0, -1):
    if n % i == 0 and prime(i):
      print(i)
      break

  break
```

```python
#W4A10

n = int(input())

ans = 0
trace = 0

for i in range(1, n+1):
  cnt = 1
  t = i
  while t > 1:
    if t % 2 == 0:
      t //= 2
    else:
      t = t * 3 + 1
    cnt += 1
  if cnt > ans:
    ans = cnt
    trace = i

print(trace, ans)
```

```python
#W4A13

def cls(a, b):
  tonga = 0
  tongb = 0
  for i in range(1, a):
    if a % i == 0:
      tonga += i
  for i in range(1, b):
    if b % i == 0:
      tongb += i
  if tonga == b and tongb == a:
    return True
  else:
    return False

a, b = map(int, input().split())

if cls(a,b):
  print("True")
else:
  print("False")
```

```python
#W4A27

def dem(chuoi):

  sotu = 0
  trongtu = False
  chuoiafter = chuoi.strip()

  for kytu in chuoiafter:
    if kytu != " ":
      if trongtu == False:
        sotu += 1
        trongtu = True
    else:
      trongtu = False

  return sotu

w = input()

print(dem(w))
```

```python
#W4A32

passw = input()

upper = False
lower = False
digit = False
alnum = False

if len(passw) >= 6:
  for kytu in passw:
    if kytu.isupper():
      upper = True
    elif kytu.islower():
      lower = True
    elif kytu.isdigit():
      digit = True
    elif not kytu.isalnum():
      alnum = True

if upper and lower and digit and alnum:
  print("Mật khẩu mạnh")
else:
  print("Mật khẩu yếu")
```
