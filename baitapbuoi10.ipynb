```python
#W9A1

def merge_sorted_list(list1, list2):
  result = []
  i = 0
  j = 0

  while i < len(list1) and j < len(list2):
    if list1[i] <= list2[j]:
      result.append(list1[i])
      i += 1
    else:
      result.append(list2[j])
      j += 1

  while i < len(list1):
    result.append(list1[i])
    i += 1

  while j < len(list2):
    result.append(list2[j])
    j += 1

  return result

list1 = list(map(int, input().split()))
list2 = list(map(int, input().split()))

print(merge_sorted_list(list1, list2))
```

```python
#W9A2

def count(matrix, target, n, m):
  cnt = 0
  row = n - 1
  col = 0
  while row >= 0 and col < m:
    if matrix[row][col] <= target:
      cnt += row + 1
      col += 1
    else:
      row -= 1
  return cnt

def kth_biggest(matrix, k):
  n = len(matrix)
  m = len(matrix[0])
  low = matrix[0][0]
  high = matrix[n - 1][m - 1]
  while low <= high:
    mid = (low + high) // 2
    if count(matrix, mid, n, m) >= k:
      high = mid - 1
    else:
      low = mid + 1
  return low

n, m = map(int, input().split)
matrix = []
for i in range(n):
  row = list(map(int, input().split()))
  matrix.append(row)
k = int(input())
print(kth_biggest(matrix, k))
```

```python
#W9A3

def swap_to_min(n):
  s = list(n)
  swapped = False
  for i in range(len(n)):
    min_i = -1
    for j in range(len(n) - 1, i, -1):
      if s[j] < s[i]:
        if i == 0 and s[j] == "0":
          continue
        if min_i == -1 or s[j] < s[min_i]:
          min_i = j
    if min_i != -1:
      s[i], s[min_i] = s[j], s[min_i]
      swapped = True
      break
  if not swapped and len(n) >= 2:
    s[-1], s[-2] = s[-2], s[-1]
  return "".join(s)

n = input()
print(swap_to_min(n))
```

```python
#W9A4

def wev(a, b):
  i = 0
  j = 0
  while i < len(a):
    if j < len(b) and a[i].upper() == b[j]:
      j += 1
    elif a[i].isupper():
      return "NO"
    i += 1
  return "YES" if j == len(b) else "NO"

k = int(input())
for i in range(1, k + 1):
  a = input().strip()
  b = input().strip()
  print(wev(a, b))
```

```python
#W9A5

def sublist(a, b):
  n = len(a)
  m = len(b)
  for i in range(m - n + 1):
    if b[i:i + n] == a:
      return "YES"
  return "NO"

a = list(input().split())
b = list(input().split())

print(sublist(a, b))
```

```python
#W9A6

def hoanvi(n, k):
  a = -1
  for j in range(k - 1, 0, -1):
    if n[j - 1] < n[j]:
      a = j - 1
      break
  if a == -1:
    return " ".join(map(str, sorted(n)))

  for j in range(k - 1, 0, -1):
    if n[j] > n[a]:
      b = j
      break
  n[a], n[b] = n[b], n[a]
  n_new = n[:a + 1] + sorted(n[a + 1:])
  return " ".join(map(str, n_new))

k = int(input())
n = list(map(int, input().split()))

print(hoanvi(n, k))
```

```python
#W9A7

def canPlaceFlowers(flowerbed, k):
  if k <= 0:
    return True
  for i in range(len(flowerbed)):
    if flowerbed[i] == 0:
      left = (i == 0) or (flowerbed[i - 1] == 0)
      right = (i == len(flowerbed) - 1) or (flowerbed[i + 1] == 0)
      if left and right:
        flowerbed[i] = 1
        k -= 1
        if k <= 0:
          return True
  return k <= 0

flowerbed = list(map(int, input().split()))
k = int(input())

print(canPlaceFlowers(flowerbed, k))
```

```python
#W9A8

def marked(moves):
  x, y = 0, 0
  s = {(0, 0)}
  for move in moves:
    if move == "L": x -= 1
    elif move == "R": x += 1
    elif move == "U": y += 1
    elif move == "D": y -= 1
    v = (x, y)
    if v in s:
      return True
    s.add(v)
  return False

moves = input()
print(marked(moves))
```

```python
#W9A10

def findRadius(houses, heaters):
  houses.sort()
  heaters.sort()
  i = 0
  r = 0
  for h in houses:
    while i < len(heaters) - 1 and abs(h - heaters[i]) >= abs(h - heaters[i + 1]):
      i += 1
    r = max(r, abs(h = heaters[i]))
  return r

houses = list(map(int, input().split))
heaters = list(map(int, input().split))
print(findRadius(houses, heaters))
```
